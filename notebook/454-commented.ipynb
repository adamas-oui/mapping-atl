```python
import folium 
import pandas as pd
```

```python
# read in csv from local directory
geodata = pd.read_csv("../data_454-3.csv", encoding='cp1252')
```

```python
# make sure city-states with small tributes are not covered by those pay more in the map
pd.options.display.max_rows = None
geodata = geodata[geodata['lon'].notnull()]
geodata = geodata[geodata['tribute_val'].notnull()]
geodata = geodata.sort_values(by=['tribute_val'],  ascending=False)
```

```python
geodata[:10]
```

```python
m_454 = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)
```

```python
for i in range(0,len(geodata)):
   folium.Circle(
      location=[geodata.iloc[i]['lat'], geodata.iloc[i]['lon']],
      popup=[geodata.iloc[i]['city_states'], geodata.iloc[i]['tribute_val']],
      radius=float(geodata.iloc[i]['tribute_val']**0.5)*300,
      color='crimson',
      fill=True
   ).add_to(m_454)
```

```python
data_432 = pd.read_csv("../data_432-1.csv", encoding='cp1252')
data_430 = pd.read_csv("../data_430-29.csv", encoding='cp1252')
```

```python
data_432 = data_432.sort_values(by=['tribute_val'],  ascending=False)
data_430 = data_430.sort_values(by=['tribute_val'],  ascending=False)
```

```python
m_432 = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)
for i in range(0,len(data_432)):
   folium.Circle(
      location=[data_432.iloc[i]['lat'], data_432.iloc[i]['lon']],
      popup=[data_432.iloc[i]['city_states'], data_432.iloc[i]['tribute_val']],
      radius=float(data_432.iloc[i]['tribute_val']**0.5)*300,
      color='crimson',
      fill=True
   ).add_to(m_432)
m_432
```

```python
m_430 = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)
for i in range(0,len(data_430)):
   folium.Circle(
      location=[data_430.iloc[i]['lat'], data_430.iloc[i]['lon']],
      popup=[data_430.iloc[i]['city_states'], data_430.iloc[i]['tribute_val']],
      radius=float(data_430.iloc[i]['tribute_val']**0.5)*300,
      color='crimson',
      fill=True
   ).add_to(m_430)
m_430
```

```python
# check which city-states have paid for both years
overlap = pd.merge(data_432, data_430, how='inner', on=['city_states'])
overlap
```

```python
m_ovlp = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)
for i in range(0,len(overlap)):
   folium.Circle(
      location=[overlap.iloc[i]['lat_x'], overlap.iloc[i]['lon_x']],
      popup=[overlap.iloc[i]['city_states'], overlap.iloc[i]['tribute_val_x']],
      radius=float(overlap.iloc[i]['tribute_val_x']**0.5)*300,
      color='crimson',
      fill=True
   ).add_to(m_ovlp)
m_ovlp
```

```python
overlap = overlap.loc[:, ['city_states', 'tribute_val_x', 'tribute_val_y']]
overlap['change'] = overlap.apply(lambda row: row.tribute_val_y - row.tribute_val_x, axis=1)
overlap
```

```python
overlap.reindex(overlap.change.abs().sort_values().index)
```

```python
# count how many city-states have changed their payments
overlap['change'].value_counts()[0]
```
